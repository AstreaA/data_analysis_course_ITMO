#Лабораторная работа №3: Кластерный анализ методом DBSCAN

```python
#  Активируем библиотеки

import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
%matplotlib inline

from sklearn import preprocessing

import os
```

```python
#  Активируем процедуру DBSCAN

from sklearn.cluster import DBSCAN
```

```python
data = './Econom_Cities_data.csv'
df = pd.read_csv(data, sep=';', decimal=',',  index_col='City')
df.head(10)
```

```python
#Имея информацию из предыдущей лабы удалим выбросы
clean_df = df.drop(['Jakarta', 'Cairo'])
```

```python
#Стандартизируем так как данные в разных диапазонах
std = preprocessing.StandardScaler()
std.fit(clean_df)
std_df = std.transform(clean_df)
std_df = pd.DataFrame(std_df, index = clean_df.index, columns = clean_df.columns)
std_df.head(10)
```

```python
# Создаем объект DBSCAN
dbscan_1 = DBSCAN(eps=0.5, metric='euclidean', min_samples=5)

# Обучим модель
dbscan_1.fit(std_df)

dbscan_1.labels_
```

```python
unique, counts = np.unique(dbscan_1.labels_, return_counts=True)
print(np.asarray((unique, counts)).T)
```

Слишком много выбросов, попробуем повысить eps = 2.

```python
dbscan_2 = DBSCAN(eps=2, metric='euclidean', min_samples=5)

dbscan_2.fit(std_df)

dbscan_2.labels_
```

```python
unique, counts = np.unique(dbscan_2.labels_, return_counts=True)
print(np.asarray((unique, counts)).T)
```

Избавились от выбросов совсем и не видим разбиение на класеры попробуем понизить eps = 1.5, также уменьшим min_samples = 4

```python
dbscan_3 = DBSCAN(eps=1.5, metric='euclidean', min_samples=4)

dbscan_3.fit(std_df)

dbscan_3.labels_
```

```python
unique, counts = np.unique(dbscan_3.labels_, return_counts=True)
print(np.asarray((unique, counts)).T)
```

Появился 1 выброс и остался один большой кластер, поэтому продолжаем понижать

```python
dbscan_4 = DBSCAN(eps=0.8, metric='euclidean', min_samples=3)

dbscan_4.fit(std_df)

dbscan_4.labels_
```

```python
unique, counts = np.unique(dbscan_4.labels_, return_counts=True)
print(np.asarray((unique, counts)).T)
```

```python
dbscan_5 = DBSCAN(eps=0.8, metric='euclidean', min_samples=2)

dbscan_5.fit(std_df)

dbscan_5.labels_
```

```python
unique, counts = np.unique(dbscan_5.labels_, return_counts=True)
print(np.asarray((unique, counts)).T)
```

Получившийся вариант выглядит наиболее правдивым, поэтому для анализа выберем его

```python
clean_df['dbscan_5'] = dbscan_5.labels_
clean_df.groupby('dbscan_5').mean()
```

Кластер -1(выброс) - люди в этих городах много работают, мало зарабатывают, относительно зарплат имеют высокие цены на "жизнь";
Кластер 0 - хороший баланс зарплат, рабочего времени и цен, относительно других городов, в этих городах в среднем получают больше, чем в остальных;
Кластер 1 - города, в которых люди много работают, мало получают, а также очень высокие цены на жизнь;
Кластер 2 - предположительно самые благоприятные города, где люди работают немного относительно других, получают одни из самых высоких зарплат(в анализируемых городах), цены на жизнь хоть и высокие, но по отношению к ЗП нормальные.
Кластер 3 - в этих городах люди работают меньше всех(анализируемых), получают среднюю зарплату, но при этом самые высокие цены на жизнь.
